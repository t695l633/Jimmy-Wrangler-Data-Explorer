I would like to do my project in the financial industry. I would like to look at the alternating prices of different currencies around the globe. For this project I have chosen 2 data sets, found below. One compares the historacle price of Euro's compared to US dollars. The second compares the price of an english Pound in comparison to US dollars. While they have the same format, the dates for the Pound chart go back much further, so I will have to join the two data sets based on date.

```python
EuroDollar = pd.read_csv('EuroDollar.csv')
```

```python
PoundDollar = pd.read_csv('PoundDollar.csv')
```

I will combine the charts using the merge function. The calling objectect, PoundDollar will be considered the "left" table, and the object passed in, EuroDollar will be considered the "right" object. The command I am calling say's that on both the right and left data sets, combine the data where the "date" column has the same value. By using the inner command I say to only combine data where the attribute value I'm searching for matches in both data sets.

```python
combinedCharts = PoundDollar.merge(EuroDollar, left_on='date', right_on='date', how='inner')
combinedCharts.head(5)
```
